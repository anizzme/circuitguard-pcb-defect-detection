```python
import cv2
import matplotlib.pyplot as plt
import os

img_path = "/content/drive/MyDrive/PCB_PROJECT/PCB_DATASET/images/Short"

img_name = os.listdir(img_path)[0]

img = cv2.imread(os.path.join(img_path, img_name))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title(img_name)
plt.axis("off")
```

```python
import cv2
import os
import matplotlib.pyplot as plt

dataset_path = "/content/drive/MyDrive/PCB_PROJECT/PCB_DATASET"

defect_path = dataset_path + "/images/Short"

img_name = os.listdir(defect_path)[0]

print("Defect Image:", img_name)

# Extract template number
template_num = img_name.split("_")[0]

template_file = template_num + ".JPG"

print("Template Image:", template_file)
```

```python
# Load defect image
defect_img = cv2.imread(os.path.join(defect_path, img_name))

# Load template image
template_path = dataset_path + "/PCB_USED"

template_img = cv2.imread(os.path.join(template_path, template_file))

# Convert to RGB for display
defect_rgb = cv2.cvtColor(defect_img, cv2.COLOR_BGR2RGB)
template_rgb = cv2.cvtColor(template_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(template_rgb)
plt.title("Template PCB")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(defect_rgb)
plt.title("Defective PCB")
plt.axis("off")
```
